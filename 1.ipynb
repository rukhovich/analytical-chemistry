```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import seaborn
```

# Задание 1

```python
a = [3.77, 16.91, 30.93, 44.03]
b = [17.29, 76.83, 134.30, 185.38]
c = [0.04, 0.16, 0.28, 0.4]
```

```python
slope, intercept, r, p, std = scipy.stats.linregress(c, a)
print "475: ", r, p, std, slope, intercept
plt.plot([0, c[-1]], [intercept, c[-1]*slope + intercept], label='475')

slope, intercept, r, p, std = scipy.stats.linregress(c, b)
print "525: ", r, p, std, slope, intercept
plt.plot([0, c[-1]], [intercept, c[-1]*slope + intercept], label='525')

plt.scatter(c, a)
plt.scatter(c, b)
plt.legend(loc=0)
plt.show()
```

#### Выбрали 425
